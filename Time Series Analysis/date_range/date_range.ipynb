### Create Date Range from date_range()

```python
rnge=pd.date_range("6/1/2020","6/30/2020",freq='B') # B refers to Business days which means it will exclude weekends. 
rnge
```

```python
#Set range as index to our data
data.set_index(rnge,inplace=True)
```

```python
data
```

```python
#Resampling data based on weekly frequency
data.Close.resample('W').mean().plot(kind='bar')
```

```python
#Indexing data from 1 June to 10 June
data['2020-06-01':'2020-06-10']
```

```python
#Change frequency from weekly to day wise
data.asfreq('D',method='pad') #padding method will carry forward prices from previous dates to the the new dates like 6,7 are not present in our initial data so it copy prices of 5th to 6th and 7th
```

```python
#Change frequency to Week wise
data.asfreq('W',method='pad')
```
