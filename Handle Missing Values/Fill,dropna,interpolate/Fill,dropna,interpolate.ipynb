# Handling Missing data using Fill, Dropna, Interpolate methods

```python
import pandas as pd
```

```python
data=pd.read_csv('weather_data.csv')
```

```python
data
```

```python
#Day column is str type
type(data.day[0])
```

```python
#Parsing day column into date type from str
data_new=pd.read_csv('weather_data.csv',parse_dates=['day'])
data_new
```

```python
type(data_new.day[0])
```

```python
#Setting index 
data_new.set_index('day',inplace=True)
data_new
```

### Fillna
function manages and let the user replace NaN values with some value of their own.

```python
#Fill missing values with 0
df1=data_new.fillna(0)
df1
```

```python
#Fill missing values with different values
df2=data_new.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'No event'
})
df2
```

```python
#Forward fill means filling missing values with next value 
df3=data_new.fillna(method='ffill')
df3
```

```python
#Forward fill means filling missing values with next value and limit=1 means only one value will be filled with next value
df4=data_new.fillna(method='ffill',limit=1)
df4
```

```python
#Backward fill means filling missing values with previous value 
df5=data_new.fillna(method='bfill')
df5
```

### Interpolate
function is basically used to fill NA values in the dataframe or series. This is a very powerful function to fill the missing values. It uses various interpolation technique to fill the missing values rather than hard-coding the value.

```python
#Using Linear interpolation with method=linear(default)
df6=data_new.interpolate()
df6
```

```python
#Using Linear interpolation with method=time
df7=data_new.interpolate(method='time')
df7
```

### Dropna
function removes all the rows which contains null values.

```python
#Removes all rows which contains atleast one null value
df9=data_new.dropna()
df9
```

```python
#Removes only those rows which has all null values
df10=data_new.dropna(how='all')
df10
```
